```python
from datasets import Dataset
dataset_tr = Dataset.from_pandas(df_tr[attributes])
dataset_val = Dataset.from_pandas(df_val[attributes])
dataset_tst = Dataset.from_pandas(df_tst[attributes])
```

```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
tokenizer = AutoTokenizer.from_pretrained("google/mt5-base")
model = AutoModelForSeq2SeqLM.from_pretrained("google/mt5-base").to(device)
```

```python
def preprocess(example):
    print(example["body"])
    print(example["supernarrative"])
    # Construct input string
    input_text = f"Clasifica urmatorul text in functie de categoria de dezinformare:\n {example['headline']} \n{example['body']} \nAutor: {example['author_name']} Media shares: {example['total_shares']}"
    model_inputs = tokenizer(input_text, max_length=1024, truncation=True, padding=True)
    labels = tokenizer(f"Clasa articolului este: {example['supernarrative']}", max_length=32, truncation=True, padding=True)
    labels[labels == tokenizer.pad_token_id] = -100
    model_inputs['labels'] = labels['input_ids']
    return model_inputs
```

```python
tokenized_dataset_tr = dataset_tr.map(preprocess, batched=False)
tokenized_dataset_val = dataset_val.map(preprocess, batched=False)
```

```python
from transformers import Trainer, Seq2SeqTrainingArguments
import numpy as np
```

```python
training_args = Seq2SeqTrainingArguments(
    output_dir="./mt5-ro-fake-news-mv2",
    per_device_train_batch_size=2,
    per_device_eval_batch_size=2,
    num_train_epochs=8,
    weight_decay=0.01,
    logging_steps=100,
    eval_steps=100, save_steps=100,
    gradient_accumulation_steps=4,
    logging_dir='./logs',
    report_to="tensorboard"
)
```

```python
def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    predictions = np.argmax(predictions, axis=1)
    return {"accuracy": (predictions == labels).mean()}
```

```python
from transformers import DataCollatorForSeq2Seq

seq2seq_data_collator = DataCollatorForSeq2Seq(tokenizer, model=model)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset_tr,
    eval_dataset=tokenized_dataset_val,
    tokenizer=tokenizer,
    data_collator=seq2seq_data_collator
)
```

```python
torch.cuda.empty_cache()
# Train model
trainer.train()

# Save model
trainer.save_model("./mt5-ro-fake-news-mv2")
tokenizer.save_pretrained("./mt5-ro-fake-news-mv2")
```

```python
# Extract loss from log history
log_history = trainer.state.log_history
train_loss = [log["loss"] for log in log_history if "loss" in log]
steps = [log["step"] for log in log_history if "loss" in log]

# Plot
plt.plot(steps, train_loss, label="Training Loss")
plt.xlabel("Steps")
plt.ylabel("Loss")
plt.title("Training Loss over Time")
plt.legend()
plt.grid()
plt.show()
```

```python
def compute_accuracy(test_set):
  total = 0;
  correct = 0;
  fn = 0;
  fp = 0;
  tp = 0;
  tn = 0;
  for row in test_set:
    print(f"ROW: {total}")
    inputs = tokenizer(f"Clasifica urmatorul text in functie de categoria de dezinformare:\n{row['headline']} \n{row['body']} \nAutor: {row['author_name']} Media shares: {row['total_shares']}",  return_tensors="pt")
    # Generate output tokens
    outputs = model.generate(**inputs, max_length=50)
    # Decode
    prediction = tokenizer.decode(outputs[0], skip_special_tokens=True)

    real_class = f"Clasa articolului este: {row['supernarrative']}"

    print("REAL: "+real_class)
    print("PREDICTED: "+prediction)

    if prediction == real_class:
      correct += 1

    if real_class == "Clasa articolului este: Fake" and prediction == "Clasa articolului este: Real":
      fp += 1
    if real_class == "Clasa articolului este: Real" and prediction == "Clasa articolului este: Fake":
      fn += 1
    if real_class == "Clasa articolului este: Real" and prediction == "Clasa articolului este: Real":
      tp += 1
    if real_class == "Clasa articolului este: Fake" and prediction == "Clasa articolului este: Fake":
      tn += 1

    total += 1

    acc = correct / total if total > 0 else 0.0
    prec = tp / (tp + fp) if (tp + fp) > 0 else 0.0
    rec = tp / (tp + fn) if (tp + fn) > 0 else 0.0
  print(f"Accuracy: {acc}, Recall: {rec}, Precision: {prec}")
  return acc, prec, rec
```

```python
tp = compute_accuracy(dataset_tst)

print(f"F1 score: {(tp[1]*tp[2]*2)/(tp[1] + tp[2]) if (tp[1] + tp[2]) > 0 else 0.0}")
```
